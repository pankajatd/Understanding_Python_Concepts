Accessing Data through Row Indexing -loc can be used to access the data through Indexing

```python
data.loc[[0,5]]## Access the 0th record   and the 5th record
```

Access aSpecific row and Specific Column

```python
data.iloc[2:4,0:2]#3 here rows(2&3 are accessed,and columns 0&1 are accessed)
```

Difference between loc and iloc:
loc: if we want to access particular rows ,we have to do the row indexing,for this loc is used
iloc:  if we want to access particular rows and cols ,we have to do row and col indexing..for this iloc is used
