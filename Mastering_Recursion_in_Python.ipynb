**Factorial of a given number**

**Every Recursive code requires Extra stack memory to store the function calls**

```python
##Function Definition
def fact(n):
  ##1. Base condition
  if(n<=1):
    return(1)
  else:
    ##2. Recursive function calls
    return(n*fact(n-1))
```

```python
##Driver function
n=6
fact_num=fact(n)
print(f"Factorial of {n} is :{fact_num}")
```

Every recursive codde has Recurrance Relation

There are 3 techniques of Recurrance relation
1.   Substitution
2.   Masters Theorem
3.   Reccursive Tree


For factorial function fact(n) ---T(n) is the Total Time Cimplexity for the entire function

Time Complexity of Base Case Condition --C(constant)

Time Complexity of Recursive Call ---T(n-1) + C ===> C--> because there is multiplication


Recurrance relation:
T(n) =T(n-1) + C

Solving this we get

T(n) =O(n)===> This is Linear Time Complexity task
Since we are using the extra Stack space
for n=5 ===> we use 5 Level Stack space,
for n=10 ===> we use 10 level stack space
Space Complexity is the extra space that is required to execute our Code
Space complexity=O(n)

def fib(n):
  ##--Base Case Condition===##
  if(n<=1):
    return(n)
    Time Complexity =C
  ##--Base Case Condition===##
  else:
  ##--Recursive Call===##
  return(fib(n-1) +fib(n-2))

Since Time Complexity Recursive Function is Exponential --Dynamic Programming came into  picture.

In Recursve Code there is **ReComputation of same function Calls**...hence the Time Complexity is Exponential ie O(2**n).. This is known by the **Overlapping Subproblem**


Whenever we see that there is **Overlapping Subproblem** we have to go with **Dynamic Programming**


**Fibbonnaci Series using Recursion**

```python
##Function Definition
def fibonnaci(n):
  ##Base Case Condition
  if n<=1:
    return(n)
  ##Recursive function call
  else:
    return fibonnaci(n-1) + fibonnaci(n-2)


##Driver Code
n=5
fib_val =fibonnaci(n)
print(f"fibonnaci of {n} is {fib_val}")

```

```python
def sumofDigits(n):
  ##Base case condition
  if n<=10:
    return(n)
  else:
    ##Recursive function call
    return ((n%10) +sumofDigits(n//10))
#Driver code
n=1234
sum_val =sumofDigits(n)
print(f"Sum of the digit{n} is {sum_val}")
```
