Create a function " calculate-total_cost(prices) to evaluate sum of the prices

```python
def calculate_total_cost(prices):
  total_cost=0
  for price in prices:
    total_cost=total_cost+price
  return total_cost
cost_prices =[10,20,30,40]
total_expend =calculate_total_cost(cost_prices)
print("The total Expenditure",total_expend)
```

Write a Function is_armstron(n) to check if the Number is Armstrong or Not

```python
def is_arm_strong(n):
  str_n=str(n)
  num_digits=len(str_n)

  ##Generator Expression
  sum_of_powers=(int(digit)**num_digits for digit in str_n)

  return   sum_of_powers==n

val =is_arm_strong(370)
```
