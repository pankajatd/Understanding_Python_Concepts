Copy in Python

1.Shallow copy
2.Deep Copy

To understand Shallow Copy..Let us consider a Nested List

```python
list3=[[1,2],[3,4]] #We say this is a 2-D lists ..since it has 2 lists.1) outer list by name list3.. This list3 contains two lists inside . which has values [1,2] and [3,4]
```

```python
list3[0] #At index -0 in llist3 we have  the list [1,2] ,hence list3[0] gives us [1,2]
```

```python
list3[0][1] #At index -1 inside the first inner list  (list[0]--> we are access the first inner list) we have 2.hence list3[0][1] gives us 2
```

```python
list3[1][1]#At index -1 inside the second inner list  (list[1]--> we are access the second inner list) we have 4.hence list3[1][1] gives us 4
```

```python
list4=list3
print(list3)
print(list4)
```

```python
##To add new element (another list) to list4.Use the function append()
list4.append([5,6])
```

```python
print(list3)
print(list4)
```

```python
print(id(list3))
print(id(list4))
```

In shallow copy the inner contents(lists) remain same or they are copied ..but the outer list is separated ,hence new obkject is created for the outer list..that is for list5.henc ethe id for list4 and list5 is different

Understanding of copy and shallow copy

```python
list5=list4.copy()## This is a shallow copy
```

```python
print(list4)
print(list5)
```

we have used list5=list4.copy() -- this means a new object is created for list5 and the contents from list4 are copied into list5.Hence we see that the id's for list4 and list5 are different.
Internal list objects are same

```python
print(id(list4))
print(id(list5))
```

```python
list5[1][0]=30
```

```python
print(list5)
print(list4)
```

```python
list5.append([7,8])
```

Since the outer lists are two different objects ..changing the contents of list5 ,dosent change the contents of list4

For Shallow Copy: Outer list is new ,Innner lists are shared.

```python
print(list5)
print(list4)
```

```python

#here we are only removing the element[30,4] from the Outer list. Hence it is not reflected in list4
list5.remove([30,4])
print(list4)
print(list5)
```
