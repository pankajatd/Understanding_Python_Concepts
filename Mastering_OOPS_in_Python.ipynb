OOPS concepts in Python

self key word helps us to Initialize the current features of the class

```python
class TechforAllWithPriya():
  ##Initialization
  #Self keyword is used to features of the class
  def __init__(self,courseID,courseName,instructorName):
    self.courseID = courseID
    self.courseName = courseName
    self.instructor = instructorName

  ##User defined function inside the class
  def course_details(self):
    return f"{self.courseID} {self.courseName}"

 ##To display complete details
  def display_details(self):
    details="Course ID: {}\nCourse Name: {}\nInstructor Name {}: ".format(self.courseID,self.courseName,self.instructor)
    return details
```

Inherriting a class
super() key word is used to inherit the feattures of the base class

```python
class Learners(TechforAllWithPriya):
  def __init__(self,courseID,courseName,instructorName,studentId,studentName):
    super().__init__(courseID,courseName,instructorName)
    self.studentId = studentId
    self.student_Name = studentName

  ##To display complete details
  def display_details(self):
    parent_details=super().display_details()
    stud_details="{}\nStudent ID: {}\nStudent Name: {}".format(parent_details,self.studentId,self.student_Name)
    return stud_details
```

```python
##Objects
##Creating an Object for the class TechforAllWithPriya
sqlcourses=TechforAllWithPriya(1,"7-Days SQL BootCamphon","Priya")
print(sqlcourses.courseID)
print(sqlcourses.courseName)
print(sqlcourses.instructor)

#calling the User_defined function inside the class
print(sqlcourses.course_details())

print("GETTING THE DETAILS WITH sqlcourses OBJECT OF TechforAllWithPriya CLASS")
##Concept of Polymorphisim
print(sqlcourses.display_details())

##Creating an Object for the class Learners
enrolledLearner=Learners(1,"7-Days SQL BootCamphon","Priya",101,"Akhil")

print("ACCESSING ALL THE DETAILS FROM BOTH PARENTS CLASS AND INHERRITED CLASS")
print(enrolledLearner.courseID)
print(enrolledLearner.courseName)
print(enrolledLearner.studentId)
print(enrolledLearner.student_Name)
print(enrolledLearner.course_details())

print("GETTING THE DETAILS WITH enrolledLearner OBJECT OF Learner CLASS")
##Concept of Polymorphisim
print(enrolledLearner.display_details())
```

Encapsulation : it is method to make the class variables private by using __

self.__instructorName.
When we define the class variables as Private..it cannot be used or modified in the other class  or outsid ethe class

```python
class Employee():
  def __init__(self,name,age):
    self.name=name
    self._age=age
```

```python
class SubEmployee(Employee):
  def show_age(self):
    print("Age : ",self._age)
```

```python
emp=SubEmployee("Priya",30)
print(emp.name)
emp.show_age()
```

Declaring protected and Private Methods

Protected Methods: Can be used within a Class or SubClass
Private Method : Can be used within a Class only

```python
class BankAccount():
  def __init__(self):
    self.balance =1000
  def _show_balance(self):##protected Method that can be used within o Class or SubClass
    print("Balance : ",self.balance)

  def __update_balance(self,amount):## Private method that can be used only within the class
    self.balance +=amount

  def deposit(self,amount):
    if amount>0:
      self.__update_balance(amount)
      self._show_balance()
```

```python
account=BankAccount()
account._show_balance()
account.deposit(500)
```
