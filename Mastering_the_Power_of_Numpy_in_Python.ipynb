Difference between the functionalities of arange() and linspace ()  in numpy

arange function - This will give the values in the matrix in thespecified range.By default values will increase by 1

```python
np.arange(1,9)
```

arange function - This will give the values in the matrix in thespecified range.Here the value is increasing  by 2

```python
np.arange(1,9,2)
```

linspace function - This function will give the values in the matrix in the specified range with even distribution .Example:(1,9,5)==> 1-start element 9-Ending range element,5--Number of elements.. linspace divides in equal space

```python
np.linspace(1,9,5)
```

Difference between shallowcopy() and deepcopy()
shallowcopy -- will point to the same address..hence the change in the original array arr2..reflects the change in the assigned array shallowArr.
deepcopy---here the newly assigned array will point to the diffferent address,hence the change in the original array is not seen in the assigned array

Ex: shallowArr=arr2--> here shallowArr and arr2 are pointing to the same addres.hence any change to the array arr2 will affect the shallowArr and the changes are seen

deepArr=np.copy(arr2)--->here deepArr and arr2 are pointing to the different addres.hence any change to the array arr2 will not affect the deepArr and hence the changes are not seen in deepArr

```python
shallowArr=arr2
```

```python
deepArr=np.copy(arr2)
```

```python
arr2[0][0] =5
```

fromfunction in numpy
Here fromfunction creates a matrix of size 4*4 using a lambda function : multiplying the row and col-index

```python
np.fromfunction(lambda i,j:i*j,(4,4))
```

fromfunction in numpy
Here fromfunction creates a matrix of size 4*4 using a lambda function : adding the row and col-index

```python
np.fromfunction(lambda i,j:i+j,(4,4))
```

fromfunction in numpy
Here fromfunction creates a matrix of size 4*4 using a lambda function : check and return True or False whenever the diagnoal is equal

```python
np.fromfunction(lambda i,j:i==j,(4,4))
```

Reshape function in Numpy

```python
dataset=np.linspace(2,9,50)#this creates a dataset with 50 elements in between 2-9 with even space
```

```python
dataset
```

```python
dataset.shape #an 1-D array is created
```

To convert the above 1-D array to 2-D array we use reshape function numpy.
Always make sure that the reshaped size of the  elemnts in the array should be equal to the original array

```python
dataset.reshape(2,1,25)  #here it is creating 2 different matrices of size (1,25)==> totally it should be equal to 50: 2*1*25=50
```
