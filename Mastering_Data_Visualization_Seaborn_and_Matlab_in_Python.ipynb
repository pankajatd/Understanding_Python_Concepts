```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
data.shape
```

Visualize the relationship between the columns 'tip' and 'total_bill'
Relation - Linear(based on the graph below)

```python
sns.relplot(x='tip',y='total_bill',data=data)
```

Based on day wise  categorization of Data is made. hue parameter is used to do this categorization 'tip' and 'total_bill' are the column considered.

```python
sns.relplot(x='tip',y='total_bill',data=data,hue  ='day')
```

```python
sns.catplot(x='day',y='tip',data=data)
```

In the above catplot we obeserve thta Sat is the Day when people give maximum tip

For catplot always use the Categorical data on the X-axis

```python
sns.catplot(x='smoker',y='tip',data=data)
```

In the above graph we obeserve that the 'Non-smoker' gives the extra tip than a 'smoker'

```python
sns.catplot(x='day',y='tip',data =data,kind='box',hue='sex')
```

from the above Graph(Box Plot) we obeserve that 'Male' give more tips

Distribution plot to see the distribution of a particular column

```python
sns.distplot(data.tip)
```

from the above graph we can see that the distribution is a "Normal Distribution"(Symmetric Distribution)

Non - Symmetric Distribution -  
*   Left skewed
*   Right skewed


*   No -Outliers then the distribution is Symmetric (Normal Distribution   -Mean=Median=Mode)
Data Imputation in a Normal Distribution,for the Missing data can be done by Mean,Median,Mode.
*   Outliers then the distribution is Non-Symmetric(Right-skewed,Left-Skewed)
Mean!=Median!=Mode)
Data Imputation for the Non-Symmetric data can be done by "Median" only because when the data is Asymmteric..it shows that there is an Outlier and if we try to Impute the Data with mean ..it comes out be a very big value ,hence data Imputation may be not correct.So hence it is done with the "Median' value
Median ==> Midlle value for the "Sorted" data
