```python
iit_shortlisted_percentage =90.0
candidate_percentage=float(input("Enter the obtained percentage: "))
```

Conditional Statements

```python
if candidate_percentage>=iit_shortlisted_percentage:
  print("Congratulations ! You are shortlisted for the IIT College")
else:
  print("Sorry ! You are not shortlisted for the IIT College")
```

Reason for using (Approach-1)if-elif-else statements to be used instead of (Approach-2) if-if-if statements.

Approach-1 is the better option ,since only 1 statement is executed.in whatever is the scienario .hence it is the optimized cod ean dExecution time is less

Approach-2 is not the better option ,since all the statements are executed even though the condition is satisfied .hence it is not the optimized version and the Execution time is more



```python
iit_shortlisted_percentage =90.0
candidate_percentage=float(input("Enter the obtained percentage: "))

if candidate_percentage>=iit_shortlisted_percentage:
  print("Congratulations ! You are shortlisted for the IIT College")
elif(candidate_percentage<iit_shortlisted_percentage and candidate_percentage>88):
  print("You are in the Waiting list")
else:
  print("Sorry ! You are not shortlisted for the IIT College")
```

**Movie Ticket Pricing**

```python
age=int(input("Enter the age of the person: "))
day=input("Is it a week Day or Weekend: ").lower()
if age<5:
  price=0
elif(age<=12):
  price =100
elif(age>=65):
    price=80
else:
    price=200
#Apply surplus charge if its a Weekend
if day=="weekend":
  price+=200
print("Movie Ticket Price:",price,"INR")
```

Check whether a String is a palindromic or not

```python
inp_str =input("Enter the string: ")
if inp_str ==inp_str[::-1]:
  print("The String is a Palindrome")
else:
  print("The String is not a Palindrome")
```

Check whether athe entered character is a Vowel or Consonant

```python
inp_chr=input("Enter the character to check if its Vowel or a Consonanat: ").lower()
if inp_chr in ['a','e','i','o','u']:
  print("Entered Character is a Vowel")
else:
  print("Entered character is a Consonant")
```
