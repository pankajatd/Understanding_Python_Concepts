Equilibrium  Point:

Equilibrium point in an Array is an Index (0-based Indexing) such that sum of all the elements before that index is same as the sum of elements after it.Return -1 if no such point exists

Example 1: Consider an example which has elements 1,2,0,3 at index (0,1,2,3) respectively

Now : considering the index  as 2:

1. add all the elements on left of index 2  ---(1+2) =3
2. add all the elements on right side of index 2 ----(3) =3

Hence the Equilibrium point =2

```python
def findEquilibrium(arr):
    # code here
    #Time Complexity O(n**2)
    for  i in range(len(arr)):
        if sum(arr[:i]) == sum(arr[i+1:]):
          return  i
    return -1
```

FInd the Second Largest Element in an Array

Merge Sort is usually used for sort() function--So the time Complexity for Sort is O(nlogn)

Time Complexity of the below code logic : O(nlogn)

```python
def findSecondLargest(arr):

  list1=list(set(arr))
  if len(list1)<2:
    return None

  list1.sort(reverse=True)
  return list1[1]


arr=[12,35,1,10,34,35]
sec_largest= findSecondLargest(arr)
print(f" The second largest is {sec_largest}")
```

FInd the Second Largest Element in an Array

Optimized Method:

Time Complexity=O(n)

```python
def findSecondLargest(arr):
  first_largest=float('-inf')
  second_largest=float('-inf')
  list1=list(set(arr))
  if len(list1)<2:
      return None
  else:
    for elem in list1:
      if elem>first_largest:
        second_largest=first_largest
        first_largest=elem
      else:
        if elem>second_largest and elem!=first_largest:
          second_largest=elem
    return second_largest

```

```python
arr=[12,35,1,10,34,35]
sec_largest= findSecondLargest(arr)
print(f" The second largest is {sec_largest}")
```
