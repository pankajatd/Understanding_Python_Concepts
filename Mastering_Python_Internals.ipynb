How is the list getting appended with the new element in the memory?

1.We know that list is a Linear Data Structure,where we can store n-number of elemnts in one go.
Mapping happens whenever a list is created and values are inserted to the list

Mapping is {key: value} pair..where key==> indicates index number , value==> value assigned at that index.

list has constant time complexity that O(1),because of "Random Access"

As the data increases the mapping gets updated.

Whenever the new elements beyond the specified length is appended
1. a new list is created with double the previous size.
Ex Initially n=10 ,10 elemnts in the list can be appended,,
Now after  that if we try to add the 11th element,a new list is created with size n*2 that is 10*20 ..Then the new elemnt is added.
2. All the previous elemnts are copied into this new list with twice the memory.

3.Once the new list is created with twice the memory , the Garbage collector clears the previous memory.
4. So again when a new elemnt needs to be appended, beyond the size of the list size, a new memory is created of size n(new value of n)*2

5. load Factor:
load factor = current Number of elements in the List/total size of the list
            =8/10 =0.8
            Load factor =0.8

            If the Load Factor is greater than or Equal to  0.8 a new memory is allocated.
