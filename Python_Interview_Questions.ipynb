5.What is difference between shallowCopy and deepCopy?

```python
##shallow Copy
import copy
list1=[[1,2,3],[4,5,6]]
```

Shalow Copy- Shallow copy just creates a view and it changes the original value list.
In shallow copy we use -copy.copy()

```python
list2=copy.copy(list1)
list2[0][2]=10
print(list1)
```

DeepCopy- Deep Copy does not modify the original list or value
In deepcopy we use -copy.deepcopy()

```python
list3=copy.deepcopy(list1)
list3[0][1]=20
print(list1)
```

8.What is list comprehension in python?
Ans : A logic or looping can be done in a single line of code.

Example for List Comprehension ;

Given a list provide the squares of every element in the list

```python
list1=[2,3,4,5,6]
list2= [x**2  for x in list1]
print(list2)
```

```python
squares=[x**2 for x in range(10)]## execute from 0 to 9
print(squares)
```
