```python
##Two Pointers Approach
#Time Complexity : O(n)
##Space Complexity -O(1) Constant

def reverse(list1):
  high=len(list1)-1
  low=0
  while(low<high):
    list1[low],list1[high]=list1[high],list1[low]
    low+=1
    high-=1
  return(list1)
```

```python
list1=[2,5,7,12,10]
rev_list=reverse(list1)
print(rev_list)
```

Create a function that accepts the list of numbers and return the sum of all the even numbers in the List

list1=[2,5,8,9]
output=2+8

```python
def list_sum(list1):
  sum=0
  for i in list1:
    if(i%2==0):
      sum+=i
  return(sum)
```

```python
list1=[12,5,8,9]
even_elem_sum=list_sum(list1)
print(even_elem_sum)
```

Create a function to calculate the fibonaci fib(5) =5

fib(5)=5

fib(0)=0
fib(1)=1
fib(2)=fib(0) +fib(1)=>0+1=1
fib(3) =fib(2) +fib(1)=>1+1=2
fib(4)=fib(3) +fib(2)=>2+1=3
fib(5)=fib(4)+fib(3) =3+2 =5




```python
def fibbonaci_num(n):
  a=0
  b=1
  for i in range(2,n+1):
    c=a+b
    a=b
    b=c
  return(c)
```

```python
fibbo_num=fibbonaci_num(5)
print(fibbo_num)
```
