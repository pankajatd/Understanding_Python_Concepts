Exception handling: When an exception occurs we do not want the Python program to stop execution... it has to handle that exception and continue further.

Below techniques help us to handle the Exceptions that occur


1.   try --Crtiticalcode is written in this block
2.   except --if an exception occurs in the try block,cod eto handle the exception is written in except block
3.   else----code to execute if no exception is raised
4.   finally---Code to execute irrespective of try block-except block has executed.
In real usecase we use the finally block to free the resources
5.raise--used to generate an exception intenionally
